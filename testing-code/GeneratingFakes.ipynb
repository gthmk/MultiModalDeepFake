```python
import requests
import os
import json
import yaml
```

```python
#import params
with open('/home/ubuntu/configs/config.yaml', 'r') as file:
    inputs = yaml.safe_load(file)
xi_api_key = inputs['eleven_labs_api_key']
```

```python
#get the voice IDs
url = 'https://api.elevenlabs.io/v1/voices'
headers = {'accept':'application/json',
           'xi-api-key':xi_api_key}
r = requests.get(url, headers=headers)
voice_data = r.json()['voices']
voice_ids = dict([(voice['name'], voice['voice_id']) for voice in voice_data])

speaker_map = {'Biden':'jb', 'Obama':'bo'}
```

```python
voice_ids
```

```python
def processWord(word):
    word = word.lower()
    return ''.join(e for e in word if e.isalnum())
```

```python
#get the transcripts from the files
path = '/Users/romitbarua/Documents/Berkeley/Spring 2023/world_leaders/combined/worldleaders20'
speakers = ['jb', 'br', 'pb', 'kh', 'cb']

data = []
data_tags = []
file_names = []

for speaker in speakers:
    for file in os.listdir(os.path.join(path,speaker)):
        file_content = []
        if ('.txt' in file) and ('phone' not in file):
            with open(os.path.join(path, speaker,file)) as f:
                lines = f.readlines()
                for line in lines:
                    
                    file_content.append(processWord(line.split(',')[0]))
                    
            data.append(file_content)
            data_tags.append(speaker)
            file_names.append(f'{speaker} - {file}')
```

```python
#convert the text into full sentences
full_text = []
for d in data:
    full_text.append(' '.join(d))    
```

```python
def generate_deepfake(voice_id, text, output_file, api_key):
    
    headers = {
        'accept': 'audio/mpeg',
        'xi-api-key': api_key,
        'Content-Type': 'application/json'
    }

    #data = '{"text": "Eleven Labs has a made a voice cloning AI that will do a good job at modeling Biden"}'
    data = f'{{"text": "{text}"}}'

    r = requests.post(f'https://api.elevenlabs.io/v1/text-to-speech/{voice_id}', headers=headers, data=data)
    
    with open(f'/Users/romitbarua/Documents/Berkeley/Spring 2023/world_leaders/ElevenLabDeepFakes/{output_file}.mpeg', 'wb') as f:
        f.write(r.content)
        f.close()
```

```python
for i, txt in enumerate(full_text):
    if data_tags[i] == 'jb':
        generate_deepfake(voice_ids['Biden'], txt, file_names[i], xi_api_key)
        
```
