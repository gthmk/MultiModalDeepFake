```python
from packages.AudioEmbeddingsManager import generateTitaNetEmbeddings

import nemo.collections.asr as nemo_asr 
import pandas as pd
speaker_model = nemo_asr.models.EncDecSpeakerLabelModel.from_pretrained(model_name='titanet_large')
```

```python
metadata = pd.read_csv('/home/ubuntu/data/wavefake_data/LJ_metadata_16000KHz.csv')
```

```python
archs = ['Full_Band_MelGan', 'HifiGan', 'MelGan', 'MelGanLarge', 'Multi_Band_MelGan', 'Parallel_WaveGan', 'Waveglow', 'ElevenLabs', 'UberDuck']
```

```python
metadata.head()
```

```python
for arch in archs:
    print(f'Currently Generating Embeddings for {arch}')
    paths = metadata[arch]
    embeddings = generateTitaNetEmbeddings(speaker_model, paths, False)
    embeddings_df = pd.DataFrame(embeddings)
    embeddings_df['path'] = paths
    embeddings_df.to_csv(f'/home/ubuntu/data/wavefake_data/Embeddings/16000KHz/{arch}.csv')
    
    
    
```
