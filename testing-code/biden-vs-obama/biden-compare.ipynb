## Load Data

```python
biden_ai = '/Users/gautham/datasets/world-leaders-dataset/biden_ai.wav'
biden_real = '/Users/gautham/datasets/world-leaders-dataset/biden_original.wav' 
```

## WIP

```python
fs_jbai, audio_jbai = wavfile.read(biden_ai)
print(f"Loaded {audio_jbai.size} samples at a sampling rate of {fs_jbai}Hz")
ipd.Audio(audio_jbai, rate=fs_jbai)
```

```python
fs_jbre, audio_jbre = wavfile.read(biden_real)
print(f"Loaded {audio_jbre.size} samples at a sampling rate of {fs_jbre}Hz")
ipd.Audio(audio_jbre, rate=fs_jbre)
```

```python
jbai, sr_jbai = librosa.load(biden_ai)
jbre, sr_jbre = librosa.load(biden_real)
```

```python
#plotting the audio array with librosa.display.wavplot
plt.figure(figsize=(14, 5))
librosa.display.waveshow(jbai, sr=sr_jbai)
```

```python
#plotting the audio array with librosa.display.wavplot
plt.figure(figsize=(14, 5))
librosa.display.waveshow(jbre, sr=sr_jbre)
```

```python
#displaying the spectrogram using librosa.display.specshow
X_jbai = librosa.stft(jbai)
Xdb_jbai = librosa.amplitude_to_db(abs(X_jbai))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb_jbai, sr=sr_jbai, x_axis='time', y_axis='hz')
```

```python
#displaying the spectrogram using librosa.display.specshow
X_jbre = librosa.stft(jbre)
Xdb_jbre = librosa.amplitude_to_db(abs(X_jbre))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb_jbre, sr=sr_jbre, x_axis='time', y_axis='hz')
```
