```python
import os
import pandas as pd
```

```python
home_path = home_path = '/home/ubuntu'
```

```python
lf_real_wav_folder = 'data/wavefake_data/LJSpeech_1.1/wavs/Original'

# 7 GANs folders
lj_full_band_mlegan_wav_folder = 'data/wavefake_data/generated_audio/ljspeech_full_band_melgan/Original'
lj_hifigan_wav_folder = 'data/wavefake_data/generated_audio/ljspeech_hifiGAN/Original'
lj_melgan_wav_folder = 'data/wavefake_data/generated_audio/ljspeech_melgan/Original'
lj_melgan_large_wav_folder = 'data/wavefake_data/generated_audio/ljspeech_melgan_large/Original'
lj_multi_band_melgan_wav_folder = 'data/wavefake_data/generated_audio/ljspeech_multi_band_melgan/Original'
lj_parallel_wavegan_wav_folder = 'data/wavefake_data/generated_audio/ljspeech_parallel_wavegan/Original'
lj_waveglow_wav_folder = 'data/wavefake_data/generated_audio/ljspeech_waveglow/Original'

folders = [lf_real_wav_folder, lj_full_band_mlegan_wav_folder, lj_hifigan_wav_folder, lj_melgan_wav_folder, lj_melgan_large_wav_folder, lj_multi_band_melgan_wav_folder, lj_parallel_wavegan_wav_folder, lj_waveglow_wav_folder]
```

```python
column_headers = ['id', 'Main_ID', 'Secondary_ID', 'Real', 'Full_Band_MelGan', 'HifiGan', 'MelGan', 'MelGanLarge', 'Multi_Band_MelGan', 'Parallel_WaveGan', 'Waveglow']
```

```python
#setup real values
ids = list(os.listdir(os.path.join(home_path, lf_real_wav_folder)))
ids = [name.replace('.wav', '') for name in ids]
ids.sort()
```

```python
metadata = pd.DataFrame(columns=column_headers)
metadata['id'] = ids
metadata['Main_ID'] = metadata['id'].str[:5]
metadata['Secondary_ID'] = metadata['id'].str[-4:]
```

```python
for idx, folder in enumerate(folders):
    files = list(os.listdir(os.path.join(home_path, folder)))
    files.sort()
    files = [os.path.join(home_path, folder, file) for file in files]
    metadata[column_headers[idx+3]] = files
    
    
```
