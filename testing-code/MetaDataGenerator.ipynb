```python
import os
import pandas as pd
```

```python
home_path = '/home/ubuntu'
sample_type = 'Original' 
```

```python
lf_real_wav_folder = f'data/wavefake_data/LJSpeech_1.1/wavs/{sample_type}'

# 7 GANs folders
lj_full_band_mlegan_wav_folder = f'data/wavefake_data/generated_audio/ljspeech_full_band_melgan/{sample_type}'
lj_hifigan_wav_folder = f'data/wavefake_data/generated_audio/ljspeech_hifiGAN/{sample_type}'
lj_melgan_wav_folder = f'data/wavefake_data/generated_audio/ljspeech_melgan/{sample_type}'
lj_melgan_large_wav_folder = f'data/wavefake_data/generated_audio/ljspeech_melgan_large/{sample_type}'
lj_multi_band_melgan_wav_folder = f'data/wavefake_data/generated_audio/ljspeech_multi_band_melgan/{sample_type}'
lj_parallel_wavegan_wav_folder = f'data/wavefake_data/generated_audio/ljspeech_parallel_wavegan/{sample_type}'
lj_waveglow_wav_folder = f'data/wavefake_data/generated_audio/ljspeech_waveglow/{sample_type}'
lj_elevenlabs_folder = f'data/wavefake_data/generated_audio/ljspeech_elevenlabs/{sample_type}'

folders = [lf_real_wav_folder, lj_full_band_mlegan_wav_folder, lj_hifigan_wav_folder, 
           lj_melgan_wav_folder, lj_melgan_large_wav_folder, lj_multi_band_melgan_wav_folder, 
           lj_parallel_wavegan_wav_folder, lj_waveglow_wav_folder, lj_elevenlabs_folder]
```

```python
column_headers = ['id', 'Main_ID', 'Secondary_ID', 'Real',
                  'Full_Band_MelGan', 'HifiGan', 'MelGan', 'MelGanLarge', 'Multi_Band_MelGan', 'Parallel_WaveGan', 'Waveglow', 'ElevenLabs']
```

```python
#setup real values
ids = list(os.listdir(os.path.join(home_path, lf_real_wav_folder)))
ids = [name.replace('.wav', '') for name in ids]
ids.sort()
```

```python
metadata = pd.DataFrame(columns=column_headers)
metadata['id'] = ids
metadata['Main_ID'] = metadata['id'].str[:5]
metadata['Secondary_ID'] = metadata['id'].str[-4:]
```

```python
def name_based_file_search(files, column):
    
    files_copy = files.copy()

    for id_name in metadata['id']:
        for file in files_copy:
            if id_name in file:
                metadata.loc[metadata['id']==id_name, column] = os.path.join(home_path, folder, file)
                files_copy.remove(file)

```

```python
for idx, folder in enumerate(folders):
    files = list(os.listdir(os.path.join(home_path, folder)))
    if len(files) == metadata.shape[0]:
        files.sort()
        files = [os.path.join(home_path, folder, file) for file in files]
        metadata[column_headers[idx+3]] = files
    else:
        name_based_file_search(files, column_headers[idx+3])
    
    
```
