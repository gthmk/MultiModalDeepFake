```python
from sklearn.decomposition import PCA
from sklearn import datasets
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
X = iris['data']
y = iris['target']
```

```python
X_train = X[:100, :]
X_test = X[100:150, :]
y_train = y[:100]
y_test = y[100:150]
```

```python
pca = PCA()
```

```python
fit_X_train = pca.fit_transform(X_train)
```

```python
pca.components_
```

```python
fit_X_train.shape
```

```python
train_loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
train_loadings
```

```python
fit_X_test = pca.transform(X_test)
```

```python
test_loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
test_loadings
```

```python
X_train_mean = X_train.mean(axis=0)
```

```python
X_train_centered = X_train - X_train_mean
```

```python
weights = np.dot(X_train_centered, pca.components_.T)
weights
```
